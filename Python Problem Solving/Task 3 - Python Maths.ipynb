Q26



You are given a complex z. Your task is to convert it to polar coordinates. 

Input Format 

A single line containing the complex number z. Note: complex() function can be used in python to convert the input as a complex number. 

Constraints
 
Given number is a valid complex number 

Output Format 

Output two lines: 

The first line should contain the value of r. 

The second line should contain the value of p. 

Sample Input 

1+2j
 

Sample Output
 
2.23606797749979 

1.1071487177940904 

Note: The output should be correct up to 3 decimal places 

```python
import cmath

z = complex(input())

r = abs(z)
p = cmath.phase(z)

print(round(r, 3))
print(round(p.real, 3))
```

Q27

ABC is a right triangle, 90 degree at B. 

Therefore, ABC = 90 degree

Point M is the midpoint of hypotenuse AC. 

You are given the lengths AB and BC. 

Your task is to find MBC (angle theta degree , as shown in the figure) in degrees. 

Input Format 

The first line contains the length of side AB. 

The second line contains the length of side BC. 

Constraints 

• 0 < AB < 100 

• 0 < BC < 100 

• Lengths AB and BC are natural numbers. 

Output Format 

Output MBC in degrees. 

Note: Round the angle to the nearest integer. 

Examples: 

If angle is 56.5000001 degree, then output 57 degree.

If angle is 56.50000000 degree, then output 57 degree.
 
If angle is 56.49999990 degree, then output 56 degree.

0 degree < theta degree< 900 

Sample Input 

10

10

Sample Output 

45 degree

```python
import math

AB = float(input())
BC = float(input())

theta = math.degrees(math.atan(AB/BC))

print(str(round(theta)) + '\xb0')
```

Q31

Integers in Python can be as big as the bytes in your machine's memory. There is no limit in size as there is: 2 to the power 31 —  1 (c++ int) or 2 to the power of 63 — 1 (C++ long long int). 

As we know, the result of a to the power of b grows really fast with increasing b. 

Let's do some calculations on very large integers. 
 
Read four numbers, a, b, c, and d, and print the result of a to the power of b + c to the power of d. 

Input Format 

Integers a, b, c, and d are given on four separate lines, respectively. 

Constraints 

1 < a <1000 

1 < b <1000 

1 < c < 1000
 
1 < d < 1000 

Output Format 

Print the result of a to the power of b +  c to the power of d on one line. 

Sample Input 

9 

29 

7

27

Sample Output 

4710194409608608369201743232

Note: This result is bigger than 2 to the power of 63 — 1. Hence, it won't fit in the long long int of C++ or a 64-bit integer. 

```python
a = int(input())
b = int(input())
c = int(input())
d = int(input())

ab = pow(a, b)
cd = pow(c, d)

result = ab + cd

print(result)
```
